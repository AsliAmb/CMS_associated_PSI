```python
import matplotlib.pyplot as plt
import seaborn as sns
from stable_event_finder import *
```

```python
'''Simulated input data contains 4000 cases and 40 alternative splicing events. From 40 events, only the first 12 are simulated 
to discriminate between CMS. The input data structure is as follows:
'''
input_dat = pd.read_csv("data/simulated_input.csv")
input_dat
```

```python
'''For our base model we are using multinomial logistic regression with L1 regularizaton and we are testing the stability of 
    event selection over 4 different strengths of regularization (c_values). The base model is fit 100 (iterations) times for each 
    bootstrapped sample wherein CMS frequencies are preserved. In the below, example we are using a selection probability cut-off (cutoff) of 
     0.1 '''

model = LogisticRegression(random_state=123, class_weight='balanced',
                   dual=False, multi_class="ovr", solver="saga", max_iter=5000, n_jobs=-1, penalty="l1")
c_values = [0.10,0.15,0.3,0.5]
iterations = 100
cutoff = 0.9
results = main("data/simulated_input.csv", c_values, model,iterations,cutoff)
```

```python
results
#As expected, features simulated to be different across (Events 1-12) have high selection probability
```

```python
#We can  look at PSI distribution of events over CMS to understand 
n_subplots = 40
for i in range(n_subplots):
    colors = {"CMS1":"#FFC300","CMS2": "#2E86C1","CMS3":"#E995DC","CMS4":"#73A482"}
    ax = sns.boxplot(x='prediction', y='Event_'+str(i+1), data=input_dat, hue="prediction", palette = colors)
    plt.legend([],[], frameon=False)
    plt.show()
 
```
