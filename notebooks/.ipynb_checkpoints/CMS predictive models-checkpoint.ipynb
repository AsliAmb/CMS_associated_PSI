```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split,GridSearchCV 
from sklearn.metrics import roc_curve, precision_recall_curve,average_precision_score, auc, recall_score, precision_score

from sklearn.metrics import confusion_matrix,roc_auc_score, plot_confusion_matrix, classification_report
from sklearn.pipeline import Pipeline 
from sklearn import metrics

from joblib import dump, load
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
"""
Converts CMS classes to numeric values
"""
def Classification(CLASS):
    if CLASS == 'CMS1':
        return 1
    elif CLASS == 'CMS2':
        return 2
    elif CLASS == 'CMS3':
        return 3
    elif CLASS == 'CMS4':
        return 4
    
"""
Prepare the input 
"""   
def prep(inputfile, TCGA=False):
    
    df=pd.read_csv(inputfile)
    df = df[df['prediction'] != 'non-concordant']
    df = df[df['prediction'] != 'Unclassified']
    df = df.dropna()
    df['prediction']=df['prediction'].apply(Classification) 
    y = df['prediction']
    if TCGA:
         x = df.drop(['prediction','case','Patient.Id'], axis=1)
    else:
        x = df.drop(['prediction','case'], axis=1)
    return(x,y)

"""
Plot multiclass roc curve
"""   
def plot_multiclass_roc(clf, X_test, y_test, n_classes,filename,figsize=(1, 1)):
    y_score = clf.predict_proba(X_test)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    all_roc = []
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    all_roc.append(roc_auc)
    print(all_roc)
    classes = ["CMS1","CMS2","CMS3","CMS4"]
    colors = ["#FFC300","#2E86C1","#E995DC","#73A482"]
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate',fontsize=17)
    ax.set_ylabel('True Positive Rate',fontsize=17)
    ax.tick_params(axis='both', which='major', labelsize=15)
    
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i],color=colors[i], label='ROC curve (area = %0.2f) for label %s' % (roc_auc[i], classes[i]))
    ax.legend(loc="lower right",prop={'size': 15})
    plt.savefig(filename, dpi=300)
    plt.show()
    
   
def calc_multiclass_roc(clf, X_test, y_test, n_classes):
   
    y_score = clf.predict_proba(X_test)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    all_roc = []

    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    all_roc.append(roc_auc)
    return(all_roc)
   
    
def disp_cm(model,x,y):
    class_names = ["CMS1","CMS2","CMS3","CMS4"] 
    disp = ConfusionMatrixDisplay.from_estimator(model,x,y, display_labels=class_names, cmap=plt.cm.Blues)
    
    
def evaluation_roc(model,x,y):
    plot_multiclass_roc(model, x, y, 4, figsize=(10, 6)) 
    disp_cm(model,x,y)
    yhat = model.predict(x)
    class_report=classification_report(y, yhat)
    print(class_report)
    
def coef (model,x_train):
    coefs = model.coef_
    genes=x_train.columns.values 
    weights = pd.DataFrame({'genes':genes, 'coef1':coefs[0], 'coef2':coefs[1],'coef3':coefs[2],'coef4':coefs[3]})
    nonzero = weights.loc[(abs(weights['coef1'])>0) | (abs(weights['coef2']) > 0) | (abs(weights['coef3'])>0) | (abs(weights['coef4']) >0)]
    return(nonzero)

def plot_multiclass_pr(clf, X_test, ytest,figsize=(1, 1)):
    precision = dict()
    recall = dict()
    average_precision = dict()
    y_score = clf.predict_proba(X_test)

    y_test_dummies = pd.get_dummies(ytest, drop_first=False).values

    classes = ["CMS1","CMS2","CMS3","CMS4"]
    colors = ["#FFC300","#2E86C1","#E995DC","#73A482"]

    for i in range(4):
        precision[i], recall[i], _ = precision_recall_curve(y_test_dummies[:, i],
                                                        y_score[:, i])
        average_precision[i] = average_precision_score(y_test_dummies[:, i], y_score[:, i])


    precision["micro"], recall["micro"], _ = precision_recall_curve(y_test_dummies.ravel(),
    y_score.ravel())
    average_precision["micro"] = average_precision_score(y_test_dummies, y_score,
                                                     average="micro")
    plt.subplots(figsize=figsize)
    for i in range(4):
        plt.plot(recall[i], precision[i],color=colors[i],
             label='Precision-recall curve of {0} (area = {1:0.2f})'
                   ''.format(classes[i], average_precision[i]))

 
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    #plt.title('Extension of Precision-Recall curve to multi-class')
    plt.legend(loc="best")
    plt.show()
```

```python
xind,yind = prep("data/PSI_model_input.csv",TCGA=False)
```

```python
#sns.set_context("talk")
x_train, x_test, y_train, y_test = train_test_split(xind,yind, test_size=0.3, random_state=132)
Cs= [0.11, 0.12,0.125, 0.13, 0.15,0.17, 0.2, 0.3, 0.4, 0.5,100]
reg_models = []
for i in range(len(Cs)):
    clf_lf_new = LogisticRegression(random_state=123,C=Cs[i], class_weight='balanced',dual=False,multi_class="ovr",solver="saga", max_iter=5000,n_jobs=-1, penalty="l1")
    lr_p = clf_lf_new.fit(x_train,y_train)
    print(coef(lr_p,x_train).shape)
    #disp_cm(lr_p,x_test,y_test)
    new_name = "auc"+str(Cs[i])+".png"
    plot_multiclass_roc(lr_p,x_test,y_test,4,new_name,figsize=(7, 5))
    plot_multiclass_pr(lr_p,x_test,y_test,figsize=(7, 5))
    reg_models.append(lr_p)
```

```python
def prep_tcga(inputfile):
    
    df=pd.read_csv(inputfile)
    df = df[df['prediction'] != 'non-concordant']
    df = df[df['prediction'] != 'Unclassified']
    #df = df.dropna()
    df['prediction']=df['prediction'].apply(Classification) 
    y = df['prediction'].values
    x = df.drop(['prediction','case','Patient.Id'], axis=1)
    return(x,y)
```

```python
xtcga, ytcga = prep("data/PSI_TCGA.csv", TCGA=True)
```

```python
k = 0
for i in reg_models:
    print(coef(i,xtcga).shape)
    disp_cm(i,xtcga,ytcga)
    new_name = "tcga_auc"+str(Cs[k]) + ".png"
    plot_multiclass_roc(i,xtcga,ytcga,4,new_name, figsize=(7, 5))
    plot_multiclass_pr(i, xtcga, ytcga,figsize=(7, 5))
    k=k+1
```
